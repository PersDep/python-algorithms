# Solution for Consecutive Summ:

Given an array of integers of size  n  and natural numbers  k  and  m , specify the minimum value of  i  for which the sum of  k + 1  consecutive elements in the array equals  m . If there is no such value, output 0. Do not use nested loops or additional arrays (solve the problem in a single pass of the original array).

### Input format

The program first receives values  n, k , and m $$( m \leqslant 10^9 ,  0 < k < n \leqslant 10^5 )$$  n is the number of elements in the array. In the next line of input, the elements of the array are provided - integers not exceeding  100  in absolute value.

### Output format

Output the answer to the problem.

```python
def find_sum_of_consecutive(n, k, target, arr):
  res = 0
  s = 0

  for i in range(k + 1): # create window for consecutive summ
    s += arr[i]

  if s != target:
    for i in range(1, n - k):
      s += arr[i + k] # add next element
      s -= arr[i - 1] # subtract first element from the original summ (deprecated)
      if s == target:
        res = i + 1
        break
  else:
    res = 1

  return res


n, k, m = map(int, input().split())
arr = list(map(int, input().split()))
print(find_sum_of_consecutive(n, k, m, arr))
```

To solve the problem optimally in a single pass through the array, we use the sliding window method. The idea of this method is to maintain the current sum of \( k + 1 \) elements and shift the window one element to the right, updating the sum.

1. Compute the sum of the first \( k + 1 \) elements of the array.
2. Traverse the array from the beginning to the end, checking at each step if the current sum equals \( m \).
3. If the current sum equals \( m \), output the index of the start of the current window and terminate the execution.
4. If the current sum does not equal \( m \), shift the window to the right: subtract the first element of the current window from the sum and add the next element (if it exists).
5. IÐ° after completing the loop we have not found a suitable window then output 0.

# Solution for Trading:

Currently, on the stock exchange, trading stocks actively employs computer systems that simplify and automate the process of buying and selling stocks. Some of them even allow trading without human intervention.

Of course, the main criterion by which such systems are evaluated is the profit generated by trading with their use. To increase it, various mathematical methods and models are applied in the development of these systems.

The main difficulty in creating such systems is that they must somehow take into account the change in the stock prices in the future and forecast it. Your task is somewhat simpler -- the selling and buying prices of stocks for the entire period of N days are already known, and it is only necessary to develop an optimal strategy for buying and selling. Moreover, for simplicity, let's assume that over these N days, you can buy stocks no more than once and sell stocks no more than once.

In addition, let's assume that buying and selling will be carried out only with stocks of one type. At the beginning of this period, you have an amount of X rubles. For each of the days, the price $$a_i$$ (ask price) at which one share can be bought, and the price $$b_i$$ (bid price) at which one share can be sold, are known. At the same time, according to the current trading rules on the exchange, it is allowed to sell and buy only a whole number of shares (for example, if you have 5 rubles and the stock price is 2 rubles, you can buy no more than two shares). You are required to write a program that, based on the available data on the stock prices for each of the days, will find the optimal strategy for buying and selling stocks.

### Input format

The first line of the input file contains two integers  N and X $$ ( 1 \leqslant N \leqslant 10^5, 1 \leqslant X \leqslant 10^6 ) $$ The second line of the input file contains  N  integers  $$ a_1, \dotsc, a_N $$ The third line of the input file contains  N  integers $$ b_1, \dotsc, b_N  ( 1 \leqslant b_i \leqslant a_i \leqslant 1000 ) $$


### Output format

In the first line of the output file, print the maximum amount you can possess at the end of the considered period. In the second line, print two numbers - the day number  $$d_1$$  when you should buy stocks, and the day number  $$d_2$$  when you should sell these stocks (the inequality  $$d_2 > d_1$$  should be satisfied). It is implied that you buy as many stocks as you can for  X  rubles, and then sell them all. If your strategy does not require buying or selling stocks, then print $$\texttt{-1 -1}$$ in the second line. If there are multiple optimal strategies, choose the interval with the smallest  $$d_1$$  and  $$d_2$$  such that  $$d_1 < d_2$$

```python
def best_strat(n, x, buy_prices, sell_prices):
    buy_date = -1
    sell_date = -1

    min_price = buy_prices[0]
    max_profit = 0


    for i in range(1, n):
        if sell_prices[i] > min_price:
            ratio = sell_prices[i] - min_price
            pot_profit = ratio * (x // min_price)
            if pot_profit > max_profit:
                max_profit = pot_profit
                sell_date = i + 1
                buy_date = buy_prices.index(min_price) + 1
        if buy_prices[i] < min_price:
            min_price = buy_prices[i]

    if not max_profit > 0:
        return x, -1, -1
    else:
        return x + max_profit, buy_date, sell_date

n, x = map(int, input().split())
buy_prices = list(map(int, input().split()))
sell_prices = list(map(int, input().split()))

max_inc, buy_date, sell_date = best_strat(n, x, buy_prices, sell_prices)
print(max_inc)
print(buy_date, sell_date)
```
